```python
import sys
import numpy as np
import matplotlib.pyplot as plt
from PyQt5.QtWidgets import QApplication, QWidget, QLabel, QLineEdit, QPushButton, QVBoxLayout, QMessageBox, QHBoxLayout, QFrame
from PyQt5.QtGui import QFont, QPalette, QColor
from PyQt5.QtCore import Qt
from matplotlib.backends.backend_qt5agg import FigureCanvasQTAgg as FigureCanvas

class QuadraticSolver(QWidget):
    def __init__(self):
        super().__init__()
        self.initUI()

    def initUI(self):
        self.setWindowTitle('Quadratic Equation Solver')
        self.setGeometry(100, 100, 500, 500)
        
        # Set background color
        palette = QPalette()
        palette.setColor(QPalette.Window, QColor(240, 240, 240))
        self.setPalette(palette)
        
        # Title Label
        self.title_label = QLabel('Quadratic Equation Solver', self)
        self.title_label.setFont(QFont('Trebuchet MS', 16, QFont.Bold))
        self.title_label.setAlignment(Qt.AlignCenter)
        
        # Input fields layout
        self.label_a = QLabel('Enter coefficient a:', self)
        self.input_a = QLineEdit(self)
        self.input_a.setText("1")
        
        self.label_b = QLabel('Enter coefficient b:', self)
        self.input_b = QLineEdit(self)
        self.input_b.setText("-3")
        
        self.label_c = QLabel('Enter coefficient c:', self)
        self.input_c = QLineEdit(self)
        self.input_c.setText("2")
        
        input_layout = QVBoxLayout()
        input_layout.addWidget(self.label_a)
        input_layout.addWidget(self.input_a)
        input_layout.addWidget(self.label_b)
        input_layout.addWidget(self.input_b)
        input_layout.addWidget(self.label_c)
        input_layout.addWidget(self.input_c)
        
        # Buttons
        self.solve_button = QPushButton('Solve', self)
        self.solve_button.setStyleSheet("background-color: #0078D7; color: white; font-size: 14px; padding: 5px; border-radius: 5px;")
        self.solve_button.clicked.connect(self.solve_quadratic)
        
        self.plot_button = QPushButton('Plot Graph', self)
        self.plot_button.setStyleSheet("background-color: #28A745; color: white; font-size: 14px; padding: 5px; border-radius: 5px;")
        self.plot_button.clicked.connect(self.plot_graph)
        
        button_layout = QHBoxLayout()
        button_layout.addWidget(self.solve_button)
        button_layout.addWidget(self.plot_button)
        
        # Separator line
        line = QFrame()
        line.setFrameShape(QFrame.HLine)
        line.setFrameShadow(QFrame.Sunken)
        
        # Matplotlib figure
        self.figure, self.ax = plt.subplots()
        self.canvas = FigureCanvas(self.figure)
        
        # Layout
        main_layout = QVBoxLayout()
        main_layout.addWidget(self.title_label)
        main_layout.addLayout(input_layout)
        main_layout.addLayout(button_layout)
        main_layout.addWidget(line)
        main_layout.addWidget(self.canvas)
        
        self.setLayout(main_layout)
        self.show()

    def solve_quadratic(self):
        try:
            a = float(self.input_a.text())
            b = float(self.input_b.text())
            c = float(self.input_c.text())
            
            discriminant = b**2 - 4*a*c
            if discriminant > 0:
                root1 = (-b + np.sqrt(discriminant)) / (2 * a)
                root2 = (-b - np.sqrt(discriminant)) / (2 * a)
                message = f'Roots are real and different:\nRoot 1: {root1}\nRoot 2: {root2}'
            elif discriminant == 0:
                root = -b / (2 * a)
                message = f'Roots are real and same:\nRoot: {root}'
            else:
                message = 'No real roots. The roots are complex numbers.'
            
            QMessageBox.information(self, 'Solution', message)
        except ValueError:
            QMessageBox.warning(self, 'Input Error', 'Please enter valid numerical values.')
    
    def plot_graph(self):
        try:
            a = float(self.input_a.text())
            b = float(self.input_b.text())
            c = float(self.input_c.text())
            
            x = np.linspace(-10, 10, 400)
            y = a*x**2 + b*x + c
            
            self.ax.clear()
            self.ax.plot(x, y, label=f'{a}x² + {b}x + {c}', color='blue')
            self.ax.axhline(0, color='black', linewidth=0.5)
            self.ax.axvline(0, color='black', linewidth=0.5)
            self.ax.grid()
            self.ax.legend()
            
            self.canvas.draw()
        except ValueError:
            QMessageBox.warning(self, 'Input Error', 'Please enter valid numerical values.')

if __name__ == '__main__':
    app = QApplication(sys.argv)
    solver = QuadraticSolver()
    sys.exit(app.exec_())
```

```python
import sys
import numpy as np
import plotly.graph_objects as go
from PyQt5.QtWidgets import QApplication, QWidget, QLabel, QLineEdit, QPushButton, QVBoxLayout, QMessageBox, QHBoxLayout, QFrame
from PyQt5.QtGui import QFont, QPalette, QColor
from PyQt5.QtCore import Qt
import tempfile
import os

# Initialize QApplication first before importing QWebEngineView
app = QApplication(sys.argv)
from PyQt5.QtWebEngineWidgets import QWebEngineView

class QuadraticGraph(QWidget):
    def __init__(self, parent=None):
        super().__init__(parent)
        self.web_view = QWebEngineView(self)
        self.setMinimumSize(500, 400)
        self.layout = QVBoxLayout()
        self.layout.addWidget(self.web_view)
        self.setLayout(self.layout)
        self.update_plot(1, -3, 2)

    def update_plot(self, a, b, c):
        x = np.linspace(-10, 10, 400)
        y = a * x ** 2 + b * x + c
        
        fig = go.Figure()
        fig.add_trace(go.Scatter(x=x, y=y, mode='lines', name=f'{a}x² + {b}x + {c}'))
        fig.update_layout(title='Quadratic Function Plot', xaxis_title='x', yaxis_title='y', template='plotly_white')
        
        html_path = os.path.join(tempfile.gettempdir(), 'plot.html')
        fig.write_html(html_path)
        self.web_view.setUrl(f'file://{html_path}')

class QuadraticSolver(QWidget):
    def __init__(self):
        super().__init__()
        self.initUI()

    def initUI(self):
        self.setWindowTitle('Quadratic Equation Solver')
        self.setGeometry(100, 100, 600, 600)
        
        palette = QPalette()
        palette.setColor(QPalette.Window, QColor(240, 240, 240))
        self.setPalette(palette)
        
        self.title_label = QLabel('Quadratic Equation Solver', self)
        self.title_label.setFont(QFont('Arial', 16, QFont.Bold))
        self.title_label.setAlignment(Qt.AlignCenter)
        
        self.label_a = QLabel('Enter coefficient a:', self)
        self.input_a = QLineEdit(self)
        self.input_a.setText("1")
        
        self.label_b = QLabel('Enter coefficient b:', self)
        self.input_b = QLineEdit(self)
        self.input_b.setText("-3")
        
        self.label_c = QLabel('Enter coefficient c:', self)
        self.input_c = QLineEdit(self)
        self.input_c.setText("2")
        
        input_layout = QVBoxLayout()
        input_layout.addWidget(self.label_a)
        input_layout.addWidget(self.input_a)
        input_layout.addWidget(self.label_b)
        input_layout.addWidget(self.input_b)
        input_layout.addWidget(self.label_c)
        input_layout.addWidget(self.input_c)
        
        self.solve_button = QPushButton('Solve', self)
        self.solve_button.setStyleSheet("background-color: #0078D7; color: white; font-size: 14px; padding: 5px; border-radius: 5px;")
        self.solve_button.clicked.connect(self.solve_quadratic)
        
        self.plot_button = QPushButton('Plot Graph', self)
        self.plot_button.setStyleSheet("background-color: #28A745; color: white; font-size: 14px; padding: 5px; border-radius: 5px;")
        self.plot_button.clicked.connect(self.update_plot)
        
        button_layout = QHBoxLayout()
        button_layout.addWidget(self.solve_button)
        button_layout.addWidget(self.plot_button)
        
        line = QFrame()
        line.setFrameShape(QFrame.HLine)
        line.setFrameShadow(QFrame.Sunken)
        
        self.graph_widget = QuadraticGraph()
        
        main_layout = QVBoxLayout()
        main_layout.addWidget(self.title_label)
        main_layout.addLayout(input_layout)
        main_layout.addLayout(button_layout)
        main_layout.addWidget(line)
        main_layout.addWidget(self.graph_widget)
        
        self.setLayout(main_layout)
        self.show()

    def solve_quadratic(self):
        try:
            a = float(self.input_a.text())
            b = float(self.input_b.text())
            c = float(self.input_c.text())
            
            discriminant = b**2 - 4*a*c
            if discriminant > 0:
                root1 = (-b + np.sqrt(discriminant)) / (2 * a)
                root2 = (-b - np.sqrt(discriminant)) / (2 * a)
                message = f'Roots are real and different:\nRoot 1: {root1}\nRoot 2: {root2}'
            elif discriminant == 0:
                root = -b / (2 * a)
                message = f'Roots are real and same:\nRoot: {root}'
            else:
                message = 'No real roots. The roots are complex numbers.'
            
            QMessageBox.information(self, 'Solution', message)
        except ValueError:
            QMessageBox.warning(self, 'Input Error', 'Please enter valid numerical values.')

    def update_plot(self):
        try:
            a = float(self.input_a.text())
            b = float(self.input_b.text())
            c = float(self.input_c.text())
            self.graph_widget.update_plot(a, b, c)
        except ValueError:
            QMessageBox.warning(self, 'Input Error', 'Please enter valid numerical values.')

if __name__ == '__main__':
    solver = QuadraticSolver()
    sys.exit(app.exec_())
```
